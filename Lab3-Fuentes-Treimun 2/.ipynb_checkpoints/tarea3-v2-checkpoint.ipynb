```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
from numpy.lib.stride_tricks import as_strided
```

```python
image = np.array(Image.open('1.png'))
```

```python
image = image / 255
row, col, _ = image.shape
print ("pixels : ", row, "*", col)
```

```python
fig = plt.figure(figsize=(15,10))
a = fig.add_subplot(1,1,1)
imgplot = plt.imshow(image)
a.set_title('imagen del perrito')
plt.show()
```

```python
image_red = image[:,:,0]
image_green = image[:,:,1]
image_blue = image[:,:,2]

L = 0.299*image_red + 0.587*image_green + 0.114*image_blue

original_bytes = image.nbytes
print("El espacio en bytes necesitado para almacenar esta imagen es", original_bytes)
```

```python
print(L[:5,0:5], "\n")
print(L[:5,5:10], "\n")
print(L[:5,10:15], "\n")
print("---------")
def separar_matrices(tamaño_matriz, L):
    m = []
    if((L.shape[0] % tamaño_matriz != 0) and(L.shape[1] % tamaño_matriz != 0)):
        print ("No se puede separar la matriz.")
    else:
        for i in range(0, L.shape[0] ,tamaño_matriz):
            for j in range(0,L.shape[1],tamaño_matriz):
                m.append(L[i:i+tamaño_matriz,j:j+tamaño_matriz])
    
        print (m[0] , "\n")
        print (m[1], "\n")
        print (m[2], "\n")
        print(len(m))
        return m
separar_matrices(5,L)
```

```python
U_r, d_r, V_r = np.linalg.svd(image_red, full_matrices = True)
U_g, d_g, V_g = np.linalg.svd(image_green, full_matrices = True)
U_b, d_b, V_b = np.linalg.svd(image_blue, full_matrices = True)
```

```python
k = 5
```

```python
U_r_k = U_r[:, 0:k]
V_r_k = V_r[0:k, :]
U_g_k = U_g[:, 0:k]
V_g_k = V_g[0:k, :]
U_b_k = U_b[:, 0:k]
V_b_k = V_b[0:k, :]

d_r_k = d_r[0:k]
d_g_k = d_g[0:k]
d_b_k = d_b[0:k]

compress_bytes = sum([matrix.nbytes for matrix in [U_r_k, d_r_k, V_r_k, U_g_k, d_g_k, V_g_k,U_b_k, d_b_k, V_b_k]])
```

```python
ratio = compress_bytes / original_bytes
print(ratio)
```

```python
image_red_approx = np.dot(U_r_k, np.dot(np.diag(d_r_k),V_r_k))
image_green_approx = np.dot(U_g_k, np.dot(np.diag(d_g_k),V_g_k))
image_blue_approx = np.dot(U_b_k, np.dot(np.diag(d_b_k),V_b_k))

image_reconstructed = np.zeros((row,col,3))

image_reconstructed[:,:,0] = image_red_approx
image_reconstructed[:,:,1] = image_green_approx
image_reconstructed[:,:,2] = image_blue_approx

image_reconstructed[image_reconstructed < 0] = 0
image_reconstructed[image_reconstructed > 1] = 1
```

```python
fig = plt.figure(figsize=(15,10))
a = fig.add_subplot(1,1,1)
imgplot = plt.imshow(image_reconstructed)
a.set_title('la carita del perrito pero en colores como le gusta a pibarra')
plt.show()
```
